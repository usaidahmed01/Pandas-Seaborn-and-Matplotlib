```python
import pandas as pd
```

```python
empDf = pd.read_excel('employee.xlsx')
empDf
```

```python
empDf.shape

empDf.head() # starting 5 elements
empDf.tail() # last 5 elements
empDf.sample(5) # random elements

empDf.info() # information of data
empDf.describe() # mean , mod , count etc

empDf.columns # columns
empDf.index # rows

empDf.memory_usage() # memory in bytes

empDf.isnull() # tells where the null values are (empDf.notnull())
empDf.isnull().sum() # count of null values

empDf.Designation # details of specific column
empDf.Designation.unique() # no repeatition
frequency = empDf.Designation.value_counts() # count of each designation
frequency

empDf.Department
empDf.Department.unique()
```

```python
import matplotlib.pyplot as plt
# matplotlib use for visualization
```

```python
plt.bar(x = frequency.index , height= frequency.values , width= 0.5) # matplotlib.pyplot usually use to make charts and graphs
# frequency.index: gives the names
# frequency.values: gives the values
```

```python
plt.pie(x = frequency.values , labels= frequency.index , autopct= '%1.1f%%' , shadow= True)
# frequency.values: gives values
# frequency.index: gives labels(names)
# autopct: gives 1 decimal value if %1.1f%%
```

```python
idFreq = empDf.Emp_ID.value_counts()
idFreq
```

```python

plt.pie(x = idFreq.values , labels= idFreq.index , autopct= lambda pct: f"{pct:.1f}%\n({int(round(pct/100.*idFreq.values.sum()))})" , shadow= True)
```
