```python
import pandas as pd

df = pd.read_excel('cleaned_employee_data.xlsx')
df
```

```python
df['Revised_Salary'] = df.Salary.apply(lambda x: x * 1.10)
df
```

```python
df['Impact_of_Revised_Salary'] = df.apply(lambda s: s['Revised_Salary'] - s['Salary'] , axis = 1) # axis = 1 cuz for each row we are checking the salary and revised salary then subtract them
df
```

```python
df[df.Impact_of_Revised_Salary == df.Impact_of_Revised_Salary.max()]
```
