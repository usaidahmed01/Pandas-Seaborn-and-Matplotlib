```python
import pandas as pd
```

```python
dfE = pd.read_excel('employee.xlsx') # pip install openpyxl for xlsx files only
dfE
```

```python
def bonus(sal):
    if sal > 100000:
        return sal * 0.1
    else:
        return sal
```

```python
dfE['bonus'] = dfE.Salary.apply(bonus) # we use apply method to apply methods on data frames
dfE
```

```python
dfE['S bonus'] = dfE.Salary.apply(lambda sal: sal* 0.2 if sal > 100000 else sal* 0.1)
dfE
```

```python
dfE['Age Bonus'] = dfE.apply(lambda row:
                                 row['Salary']* .5 if row['Age'] >= 50
                                 else row['Salary']* .4 if row['Age'] >= 40
                                 else row['Salary']* .1 , axis = 1) # axis = 1 because means row by row if axis = 0 columns by columns
dfE
```

```python
# to REMOVE temporarily The column we use .drop

dfE.drop('S bonus' , axis = 1)

# Permanant Delete: 
# del dfE['S bonus'] 
```
